```python
import torch

from pcdet.config import cfg, cfg_from_yaml_file
from pcdet.datasets import build_dataloader
from pcdet.utils import common_utils

cfg_from_yaml_file('/home/eslim/3D_tracking/tools/cfgs/nuscenes_models/pv_rcnn_plusplus_resnet.yaml',cfg)
# cfg.EXP_GROUP_PATH = '/'.join('second_test.yaml'.split('/')[1:-1])

logger = common_utils.create_logger('/home/eslim/log.txt', rank=1)

dataset, _, _ = build_dataloader(
        dataset_cfg=cfg.DATA_CONFIG,
        class_names=cfg.CLASS_NAMES,
        batch_size=1,
        dist=False, workers=1, logger=logger, training=False
    )

voxels = dataset[5]['voxels']
voxel_num_points = dataset[5]['voxel_num_points']
voxels = torch.from_numpy(voxels)
voxel_num_points = torch.from_numpy(voxel_num_points)


invalid_id=[]

for i in range(voxels.size(0)):
    near_zero_pts = 0
    for _,val in enumerate(voxels[i]):
        if abs(val[:3].sum())<1e-3:
            near_zero_pts+=1

    if voxel_num_points[i]!=(voxels.size(1)-near_zero_pts):
        invalid_id.append(i)
```
